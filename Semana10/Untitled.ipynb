## **Utilizamos metrópolis para encontrar la función que mejor se ajuste**

Variamos los parámetros:
omega de materia entre 0 y 1.
bias entre 0.1 a 5

**Un camino**

```python
cosmo = cosmology.setCosmology('planck15')
print(cosmo)

def Pk_Om(Om_,b,k):
    z =0.57
    cosmo.Om0 = Om_
    return b**2*cosmo.matterPowerSpectrum(k,z)

def likelihood(Om, b, data):
    equis = data[0]
    ye = data[1]
    yerr = data[2] 
    model = Pk_Om(Om, b, equis)
    likelihood_0 = (ye-model)**2/(yerr**2)
    return likelihood_0.sum()
```

```python
x = pk_cmasdr12[0][:15]
y = pk_cmasdr12[1][:15]
yerror = pk_cmasdr12[2][:15]

rango = 1500

Om_0 = 0.2
b0 = 1.0
lnlike0 = -1*likelihood(Om_0, b0, [x,y,yerror])

Om_sample = [Om_0]
b_sample = [b0]

sigma = 0.05

for j in range(1,rango):
    Om_new = np.random.normal(Om_sample[len(Om_sample)-1], sigma)
    b_new = np.random.normal(b_sample[len(b_sample)-1], sigma)
    if Om_new <= 1.0 and Om_new >= 0.1 and b_new <= 4 and b_new >= 0.1:
        like = -1*likelihood(Om_new, b_new, [x,y,yerror])
        if like>lnlike0:
            lnlike0 = like
            Om_sample.append(Om_new)
            b_sample.append(b_new)
        elif lnlike0-like<-math.log(random.uniform(0,1)):
            lnlike0 = like
            Om_sample.append(Om_new)
            b_sample.append(b_new)
```

```python
plt.figure(figsize = (6,6))
plt.plot(Om_sample,b_sample, 'o', markersize = 2)
plt.xlabel('Om')
plt.ylabel('b')
plt.title('Camino seguido para llegar a $Om$ y $b$')
plt.grid('True')
```
