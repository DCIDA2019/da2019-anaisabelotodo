Importamos librerías. 

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

plt.style.use('seaborn')
```

```python
from colossus.cosmology import cosmology
```

```python
cosmo = cosmology.setCosmology('planck15')
print(cosmo)
```

Descargamos datos:

```python
pk_cmasdr12 = np.loadtxt('Semana13/GilMarin_2016_CMASSDR12_measurement_monopole_post_recon.txt').T
```

------------------------------

### **Dos parámetros**

**Graficamos con los valores de los parámetros que nos da Multinest.**

```python
plt.figure(figsize = (7,7))
plt.loglog()

k = 10**np.linspace(-5,6,100000)
cosmo.Om0 = 0.357
Pk = 2.141**2*cosmo.matterPowerSpectrum(k,0.57)

plt.plot(k,Pk, '-', label = 'Om0 = 0.357, b = 2.141') # Con bias
plt.errorbar(pk_cmasdr12[0], pk_cmasdr12[1], yerr = pk_cmasdr12[2], fmt = '.b')
  
plt.xlim(1e-3,1)
plt.ylim(100,3e6)
plt.legend()
plt.xlabel('k(Mpc/h)')
plt.ylabel('P(k)')
```

------------------------------------------------------

### **Tres parámetros**

Graficamos con los valores que Multinest encontró:

```python
plt.figure(figsize = (7,7))
plt.loglog()

k = 10**np.linspace(-5,6,100000)
cosmo.Om0 = 0.357
Pk = (1.85)**2*(1+0.36)*cosmo.matterPowerSpectrum(k,0.57)

plt.plot(k,Pk, '-', label = 'Om0 = 0.357, b = 1.85, beta = 0.36') # Con bias
plt.errorbar(pk_cmasdr12[0], pk_cmasdr12[1], yerr = pk_cmasdr12[2], fmt = '.b')
  
plt.xlim(1e-3,1)
plt.ylim(100,3e6)
plt.legend()
plt.xlabel('k(Mpc/h)')
plt.ylabel('P(k)')
```
