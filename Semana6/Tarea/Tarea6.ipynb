```python
from astropy.io import fits

import math
import os
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
from astropy.table import Table
```

```python
datos = Table.read('survey-bricks-dr8-north.fits')
```

```python
datos[0:5]
```

## **DR8**

```python
ra = datos['ra']
dec = datos['dec']
```

```python
plt.scatter(ra,dec)
```

```python
exp_g = datos['nexp_g']
max(exp_g)
```

**Graficamos y hacemos barra de color que depende del número de exposiciones en g**

```python
plt.style.use('classic')

ra = datos['ra']
dec = datos['dec']
exp_g = datos['nexp_g']

plt.scatter(ra, dec, label=None,
            c=exp_g, cmap='viridis',linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('RA (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label='Number of exposures in g')
plt.clim(0, max(exp_g))
plt.grid('True')

plt.title('DR8 Number of exposures in g');
```

Notamos que no es muy visible la dfierencia del número de exposiciones en g de cada punto. Graficamos de nuevo. 

```python
plt.style.use('classic')

ra = datos['ra']
dec = datos['dec']
exp_g = datos['nexp_g']

plt.scatter(ra, dec, label=None,
            c=exp_g, s = 1, cmap='viridis',linewidth=0, alpha=1)
plt.axis(aspect='equal')
plt.xlabel('RA (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label='Number of exposures in g')
plt.clim(0, 7)
plt.grid('True')

plt.title('DR8 Number of exposures in g');
```
