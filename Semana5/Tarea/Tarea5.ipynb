## **The select statement**

Lo más básico de SQL es el select statement, que es donde podemos seleccionar nuestra base datos y especificar qué columnas quiero que aparezcan (segundo argumento)

```python
pdsql("SELECT * FROM lluvias_2019;") # * Para que aparezcan todas las columnas
```

SELECT y FROM son clases para filtrar datos. 

```python
pdsql("SELECT LON, LAT FROM lluvias_2019") # Para mostrar sólo longitud y latitud
```

También filtramos con la clase WHERE. Vea el ejemplo siguiente:

```python
pdsql("SELECT LON, LAT FROM lluvias_2019 WHERE ENE = 0")
```

Con la clase ORDER BY ordenamos con respecto a una columna, ya sea de menor a mayor o en orden alfabético.

```python
pdsql("""SELECT LON, LAT 
    FROM lluvias_2019
    -- WHERE ENE = 0
    ORDER BY LAT""")

# -- se toma como comentario
```

Nota: FROM, WHERE, ORDER BY son clauses opcionales y tienen un orden en la sintaxis.

### **The select clause**

Podemos aparecer columnas en el orden que queramos.

```python
pdsql("""SELECT LON, LAT, NOMBRE, EDO
    FROM lluvias_2019;""")
```

Podemos hacer operaciones aritméticas y renombrar columnas utilizando AS.

```python
pdsql("""SELECT 
    LON, 
    LAT, 
    EST, 
    Tmed AS 'T (°C)', 
    Tmed + 273.15 AS 'T (K)',
    (Tmed*9)/5 + 32 AS 'T (°F)' -- Renombramos columnas con AS
    FROM temp_media_2018""")
```

Si tenemos estados que se repiten en los datos de temperaturas, utilizamos DISTINCT para obtener una tabla donde sólo aparezcan los estados que no se repiten.

```python
pdsql("""SELECT DISTINCT EDO
    FROM lluvias_2019""")
```

### **The AND, OR and NOT operators**

En el siguiente ejemplo combinaremos condiciones múltiples para filtrar datos. 

```python
pdsql("""SELECT Lon, Lat, Edo, Est, Tmed 
    FROM temp_media_2018
    WHERE Tmed >= 16 AND Edo = 'TAB'""") # UTilizamos AND para hacer la condición múltiple
```

Utilizando AND, OR, NOT:

```python
pdsql("""SELECT Lon, Lat, Edo, Est, Tmed 
    FROM temp_media_2018
    WHERE NOT (Tmed >= 16 OR Edo = 'TAB') AND Lat >30""") # Se filtran los datos donde Tmed 
#es mayor que 16 o que pertenecen al estado de Tabasco. 
# Usamos paréntesis para indicar la jerarquía en las condiciones múltiples.
```

### **The IN operator**

Con IN podemos resumir múltiples condiciones al filtrar datos. 

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Edo = 'TAB' OR Edo = 'CHI' OR Edo = 'GTO'""");

#En lugar de combinar múltiples condiciones hacemos algo más sencillo
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Edo IN ('TAB', 'CHI','GTO')""")
```

Y si en lugar de mostrar los datos de Tabasco, Chihuahua y Guanajuato queremos quitarlos, utilizamos NOT.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Edo NOT IN ('TAB', 'CHI','GTO')""")
```

### **The BETWEEN operator**

Utilizamos BETWEEN para filtrar las temperaturas entre 10 y 15 grados centígrados.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Tmed>= 10 and Tmed<=15""");

# Se reescribe la misma instrucción con BETWEEN
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Tmed BETWEEN 10 AND 15""")
```

### **The LIKE operator**

Con LIKE podemos filtrar los estados que empiezan con la letra G. Nota: Utilizamos % para indicar que hay otros caracteres que ignoramos después de la G.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Edo LIKE 'G%'""")
```

Si queremos los estados que tengan una R en el medio de su nombre usamos LIKE '%R%' y utilizamos AND para filtrar los estados que tengan como segunda letra la C, donde nuevamente usamos LIKE pero ahora con '_C'.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Edo LIKE '%C%' AND Edo LIKE '_C'""") # Para mostrar datos donde los estados tienen una C y su segunda letra es C
```

Para filtrar los datos donde el estado empieza con G y termina en O usamos LIKE 'G_O'.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Edo LIKE 'G_O'""")
```

### **The IS NULL Operator**

Para buscar datos que no tienen atributos utilizamos IS NULL

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    WHERE Tmed IS NULL
    """)
```

La tabla está vacía debido a que ninguno de los datos tiene un dato de tipo NULL.

### **The ORDER BY clause**

PAra ordenar los estados de forma ascendente utilizamos ORDER BY

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    ORDER BY Edo""")
```

Para ordenar de forma descendente utilizamos DESC.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    ORDER BY Edo DESC""")
```

Si queremos ordenar por estado de forma descendente y, a la vez, por latitud de forma ascendente, hacemos lo siguiente:

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    ORDER BY Edo DESC, Lat""")
```

En el siguiente ejemplo mostramos sólo la columna de longitud, latitud, Edo, Est y Tmed, pero ordenamos primero por longitud y luego por latitud. 

```python
pdsql("""SELECT Lon, Lat, Edo, Est, Tmed
    FROM temp_media_2018
    ORDER BY 1, 2""")
```

### **The LIMIT clause**

Para mostrar sólo cierta cantidad de datos utilizamos LIMIT.

```python
pdsql("""SELECT * -- Seleccionamos todas las columnas
    FROM temp_media_2018
    LIMIT 400,10 -- Empieza en el 400 y toma 10 datos después del 400.
    """)
```

### **Inner Joins**

Cuando queremos reunir datos de dos data frames diferentes, podemos hacer uso de inner join. En el siguiente ejemplo, se usa JOIN para unir los datos de temp_media_2018 y temp_media_2019 en ciudades que sólo los dos tienen en común. En el query se muestra la tabla con los datos de los dos data frames.

```python
pdsql("""SELECT o.Lon, o.Lat, 
    o.Edo,
    o.Est,
    o.Tmed AS 'Tmed 2018',
    n.Tmed AS 'Tmed 2019'
    FROM temp_media_2018 o -- Ponemos un alias a las temperaturas del 2018
    JOIN temp_media_2019 n -- Ponemos un alias a las temperaturas del 2019 para embellecer el código 
    ON o.Est = n.Est  """) # Inner es opcional, así que puede no escribirse
```
