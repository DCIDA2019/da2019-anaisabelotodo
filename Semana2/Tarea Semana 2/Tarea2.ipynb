```python
df.mean()
```

Incluso el promedio entre Tmin, Tmed y Tmax

```python
df.describe()
```

En esta sección también aprendemos a generar DataFrameGroupBy objects. Con este tipo de objetos no sólo dividimos el data frame en otros aparte, sino que podemos realizar operaciones sólo para esos objetos. 

```python
df.groupby('Tmin')
df.groupby('Tmin').sum() # Primero se suman los valores en índices repetidos
# luego se ordenan del menor al mayor los valores en Tmin, y eso cambia todo el df
```

```python
df.groupby('Est').aggregate(['min', np.median, max])
```

### - Pivot tables

```python
fare = pd.qcut(df['Tmed'], 3) # Se divide en 3 rangos
df.pivot_table('Tmed', ['Est', age], [fare, 'Tmin'])
```

### - High-Performance Pandas: eval() and query()

```python
tempm2019 = pd.Series(temp_media_2019.Tmed[0:100])
tempm2019.index = [temp_media_2019.Est[0:100]]
tempma2019= pd.Series(temp_maxima_2019.Tmax[0:100])
tempma2019.index = [temp_maxima_2019.Est[0:100]] 
tempmi2019= pd.Series(temp_minima_2019.Tmin[0:100])
tempmi2019.index = [temp_minima_2019.Est[0:100]]
```

```python
df1,df2,df3 = pd.DataFrame({'Tmin':tempmi2019, 'Tmed': tempm2019, 'Tmax': tempma2019})
```

Calculamos el tiempo que tardamos en generar los 3 data frames

```python
%timeit df1 + df2 + df3
```

En pandas hay una forma más eficiente de generar lo anterior. Se utiliza eval y en éste se utilizan expresiones de cadenas para eficientar operaciones computacionales al utilizar Data Frames.

```python
%timeit pd.eval('df1 + df2 + df3')
```

eval() utiliza menos memoria y genera los data frames de una forma más rápida. 

query() también se basa en evaluar cadenas.

```python
temp_minima_2019 = pd.read_csv('tempmi2019.csv', usecols = (4,5), encoding='latin-1')
temp_media_2019 = pd.read_csv('tempm2019.csv', usecols = (4,5), encoding='latin-1')
temp_maxima_2019 = pd.read_csv('tempma2019.csv', usecols = (4,5), encoding='latin-1')
lluvias_2019 = pd.read_csv('lluvia2019.csv', usecols = (4,5), encoding='latin-1')

temp_minima_2018 = pd.read_csv('tempmi2018.csv',usecols = (4,5),encoding='latin-1')
temp_media_2018 = pd.read_csv('tempm2018.csv',usecols = (4,5),encoding='latin-1')
temp_maxima_2018 = pd.read_csv('tempma2018.csv',usecols = (4,5),encoding='latin-1')
lluvias_2018 = pd.read_csv('lluvias2018.csv',usecols = (4,5),encoding='latin-1')
```

### - Para las temperaturas

Utilizamos multiIndex para que en la tabla de datos se observe una diferencia entre 2018 y 2019 al presentar las temperaturas mínimas, medias y altas en cada estado.

```python
temp_media_2018.index
```

1099 es el número más grande de índices en los archivos (el archivo de temperaturas del 2018 tiene 1099 ciudades), es por eso que si en los demás archivos del 2019, por ejemplo, debemos verificar que la lista generada al leer el archivo contenga ese mismo número de ciudades. Si no las tiene, al leer la componente i de la lista de temperaturas del 2019 se generará un error. Es por eso que se hizo un try y un except, para que en caso que no exista el elemento i de la lista, este dato se tome como 'Nan'.

```python
# En index guardamos las ciudades y en temperaturas guardamos las temperaturas de cada ciudad
index = []
temperaturas = []
for i in range(1099):
    index.append((temp_minima_2018.Est[i], '2018')) # Clasificamos ciudades en 2018 y 2019 
    try:
        index.append((temp_minima_2019.Est[i], '2019'))
    except:
        index.append(('Nan', '2019'))    #Cambiamos por nan en caso de que no exista el elemento i
    try:
        temperaturas.append((temp_minima_2018.Tmin[i]))
    except:
        temperaturas.append(('Nan'))
    try:
        temperaturas.append((temp_minima_2019.Tmin[i]))
    except:
        temperaturas.append(('Nan'))
```

```python
# Generamos un object series con index y temperaturas
pop = pd.Series(temperaturas, index=index)    
```

```python
# Generamos un multiIndex para hacer distinción entre 2018 y 2019
index = pd.MultiIndex.from_tuples(index)
index
```

```python
# Con esto reindexamos el Object series llamado pop. Puede observarse las ciudades y si tienen datos en 2018 o 2019 en temperatura
pop = pop.reindex(index)
pop
```

```python
# Anteriormente habíamos guardado datos de Tmin para 2018 y 2019
# Para anexar las temperaturas media y máxima de 2018 y 2019 generamos dos listas de datos
# Nuevamente tenemos que verificar si existe el elemento i de la lista
lista1 = []
lista2 = []
for i in range(1099):
    try:
        lista1.append((temp_media_2018.Tmed[i]))
    except:
        lista1.append(('Nan'))
    try:
        lista1.append((temp_media_2019.Tmed[i]))
    except:
        lista1.append(('Nan'))
    try:
        lista2.append((temp_maxima_2018.Tmax[i]))
    except:
        lista2.append(('Nan'))
    try:    
        lista2.append((temp_maxima_2019.Tmax[i]))
    except:
        lista2.append(('Nan'))
```

```python
# Finalmente generamos el data frame de los datos. Pop contiene los datos de Tmin, pero
# en lista 1 y lista 2 están los datos de Tmed y Tmax, respectivamente
pop_df = pd.DataFrame({'Tmin': pop,
                       'Tmed': lista1,
                      'Tmax': lista2})
pop_df
```

### - Para las lluvias

Para las lluvias es un poco más sencillo, ya que sólo son dos columnas de datos para cada ciudad, la precipitación en 2018 y la precipitación en 2019.
